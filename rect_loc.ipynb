## 定位发票中的识别区域


```python
from utils.rectangle import get_rects, is_intersect, solve_conincide

import cv2
import matplotlib.pyplot as plt
```

```python
def show_img(img, cmap=None):
    plt.figure(figsize=(15, 7))
    if cmap:
        plt.imshow(img, cmap=cmap)
    else:
        plt.imshow(img)
    plt.show()
```

```python
invoice_img = cv2.imread('./data/test4.jpg')
invoice_img = cv2.cvtColor(invoice_img, cv2.COLOR_RGB2BGR)

show_img(invoice_img)
```

```python
rects = get_rects(invoice_img)
```

```python
boxs_img = invoice_img.copy()

for i, (x, y, w, h) in enumerate(rects):
    cv2.rectangle(boxs_img, (x, y), (x+w, y+h), (0, 0, 255), 1)
    cv2.putText(boxs_img, str(i + 1), (int(x+w/2), int(y+h/2)), cv2.FONT_HERSHEY_PLAIN, 1, (255, 0, 0), 1)
    
show_img(boxs_img)
```

```python
boxs_img = cv2.cvtColor(boxs_img, cv2.COLOR_RGB2BGR)
cv2.imwrite('./data/invoice_box_num.jpg', boxs_img)    
```
